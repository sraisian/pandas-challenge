```python
#import packages
import pandas as pd
import csv
import os
import numpy as np
```

```python
#read csv file
datafile = "Resources/HeroesofPymoli_purchase_data.csv"
#import csv file into df 
purchdata= pd.read_csv(datafile)
purchdata.head()
```

```python
#1) Player Count 
#1.1) Display total number of players
totalplayers = purchdata["SN"].nunique()
totaldf = {"Total Players": [totalplayers]}
totaldf = pd.DataFrame(totaldf)
totaldf
```

```python
#2)Purchasing Analysis (Total)
#2.1) Number of Unique items 
uniqueitems = purchdata["Item ID"].nunique()
#2.2) Average Purchase Price
avgprice = round(purchdata["Price"].mean(), 2)
#2.3) Total Number of Purchases 
totalpurch = purchdata["Purchase ID"].count()
#2.4) Total Revenue 
totalrev = purchdata["Price"].sum()
```

```python
# 2.5) Create summary data frame to hold above results
purchanalysis = {"Number of Unique Items": [uniqueitems], "Average Price": [avgprice], "Number of Purchases": [totalpurch], "Total Revenue": [totalrev]}
purchdf = pd.DataFrame(purchanalysis)
# Optional: give cleaner formatting
purchdf["Average Price"] = purchdf["Average Price"].map("${:.2f}".format)
purchdf["Total Revenue"] = purchdf["Total Revenue"].map("${:,.2f}".format)
# Display summary data frame
purchdf
```

```python
#3) Gender Demographics 
# Group by gender, show total count and percentage of players
#total count 
genderdemo = purchdata.groupby(["Gender"]).nunique()
genderdf = genderdemo["SN"]
genderdf = pd.DataFrame(genderdf)
genderdf = genderdf.rename(columns = {"SN": "Total Count"})
genderdf
#3.1) Percentage and Count of Male Players
# Percentage
genderdf["Percentage"] = genderdf["Total Count"]/genderdf["Total Count"].sum()
genderdf["Percentage"] = genderdf["Percentage"].map("{:,.2%}".format)
#Display Summary Table
genderdf
```

```python
#4) Purchasing Analysis (Gender)
# Group by Gender and get following:
gendf = purchdata.groupby(["Gender"]).nunique()
# Purchase Count 
genpurchdf = gendf["Purchase ID"]
genpurchdf = pd.DataFrame(genpurchdf)
genpurchdf = genpurchdf.rename(columns = {"Purchase ID": "Purchase Count"})
genpurchdf
# Average Purchase Price 
genderdf2 = purchdata.groupby("Gender")
genpurchdf["Average Purchase Price"] = genderdf2["Price"].mean()
# Total Purchase Value 
genpurchdf["Total Purchase Value"] = genderdf2["Price"].sum()
# Average Purchase Total per Person 
genpurchdf["Avg Total Purchase per Person"] = genpurchdf["Total Purchase Value"]/genderdf["Total Count"]
# Reformatting and styling 
genpurchdf["Avg Total Purchase per Person"] = genpurchdf["Avg Total Purchase per Person"].map("${:.2f}".format)
genpurchdf["Total Purchase Value"] = genpurchdf["Total Purchase Value"].map("${:,.2f}".format)
genpurchdf["Average Purchase Price"] = genpurchdf["Average Purchase Price"].map("${:.2f}".format)
#display summary table
genpurchdf
```

```python
## 5) Age Demographics

# Establish bins for ages 
agebins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
agenames = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
agedf = pd.DataFrame(purchdata)
# Categorize the existing players using the age bins. 
# Hint: use pd.cut()
agedf["Age Summary"] = pd.cut(purchdata["Age"], agebins, labels = agenames)
agegroupdf = agedf.groupby("Age Summary")

# Calculate the numbers and percentages by age group
totalcount = agegroupdf["SN"].nunique()
totalper = totalcount/totalcount.sum()
# Create a summary data frame to hold the results
agedemodf = pd.DataFrame({"Total Count": totalcount, "Percentage of Players": totalper})
agedemodf
# Optional: round the percentage column to two decimal points
agedemodf["Percentage of Players"] = agedemodf["Percentage of Players"].map("{:,.2%}".format)
# Display Age Demographics Table
agedemodf
```

```python
# 6) Purchase Analysis (Age)
#Bin the purchase_data data frame by age
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
purchcount = agegroupdf["Purchase ID"].count()
avgpurch = agegroupdf["Price"].mean()
totalpurchvalue = agegroupdf["Price"].sum()
avgtotalper = totalpurchvalue/purchcount

#Create a summary data frame to hold the results
agedemodf2 = pd.DataFrame({"Purchase Count": purchcount, "Average Purchase Price": avgpurch, "Total Purchase Value": totalpurchvalue, "Avg Total Purchase per Person": avgtotalper})
#Optional: give the displayed data cleaner formatting
agedemodf2["Average Purchase Price"] = agedemodf2["Average Purchase Price"].map("${:,.2f}".format)
agedemodf2["Total Purchase Value"] = agedemodf2["Total Purchase Value"].map("${:,.2f}".format)
agedemodf2["Avg Total Purchase per Person"] = agedemodf2["Avg Total Purchase per Person"].map("${:,.2f}".format)
#Display the summary data frame
agedemodf2
```

```python
# 7) Top Spenders 
#Run basic calculations to obtain the results in the table below
topspenddf = purchdata.groupby("SN")
toppurchcount = topspenddf["Purchase ID"].count()
topavgpurchprice = topspenddf["Price"].mean()
toptotalpurch = topspenddf["Price"].sum()

#Create a summary data frame to hold the results
topspenders = pd.DataFrame({"Purchase Count": toppurchcount, "Average Purchase Price": topavgpurchprice, "Total Purchase Value": toptotalpurch})
#Sort the total purchase value column in descending order
topspenders = topspenders.sort_values (["Total Purchase Value"], ascending = False).head()
topspenders
#Optional: give the displayed data cleaner formatting
topspenders["Total Purchase Value"] = topspenders["Total Purchase Value"].map("${:,.2f}".format)
topspenders["Average Purchase Price"] = topspenders["Average Purchase Price"].map("${:,.2f}".format)
#Display a preview of the summary data frame
topspenders
```

```python
#8) Most Popular Items - 
# Identify the 5 most popular items by purchase count, then list (in a table):
 
#Retrieve the Item ID, Item Name, and Item Price columns
items = purchdata[["Item ID", "Item Name", "Price"]]

#Group by Item ID and Item Name. 
mostpopitem = items.groupby(["Item ID", "Item Name"])

#Perform calculations to obtain purchase count, item price, and total purchase value
purchasecount = mostpopitem["Item ID"].count()
itemprice = mostpopitem["Price"].mean()
totalvalue = mostpopitem["Price"].sum()

#Create a summary data frame to hold the results
mostpopular = pd.DataFrame({"Purchase Count": purchasecount, "Item Price": itemprice, "Total Purchase Value":totalvalue})

#Sort the purchase count column in descending order
mostpopular2 = mostpopular.sort_values (["Purchase Count"], ascending = False).head()

#Optional: give the displayed data cleaner formatting
mostpopular2["Total Purchase Value"] = mostpopular2["Total Purchase Value"].map("${:,.2f}".format)
mostpopular2["Item Price"] = mostpopular2["Item Price"].map("${:,.2f}".format)

#Display a preview of the summary data frame
mostpopular2
```

```python
#9) Most Profitable Items - 
# Identify the 5 most profitable items by total purchase value, then list (in a table): 
#Sort the above table by total purchase value in descending order
mostprofitable = mostpopular.sort_values (["Total Purchase Value"], ascending = False).head()


#Optional: give the displayed data cleaner formatting
mostprofitable["Total Purchase Value"] = mostprofitable["Total Purchase Value"].map("${:,.2f}".format)
mostprofitable["Item Price"] = mostprofitable["Item Price"].map("${:,.2f}".format)

#Display a preview of the data frame
mostprofitable
```
